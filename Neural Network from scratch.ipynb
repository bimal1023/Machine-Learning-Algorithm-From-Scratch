# Neural network from Scratch

# Import numpy

```python
import numpy as np 
```

# Create neural network class


```python
class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size, lr=0.1):
        # weights and biases
        self.W1 = np.random.randn(input_size, hidden_size) * 0.1
        self.b1 = np.zeros((1, hidden_size))
        
        self.W2 = np.random.randn(hidden_size, output_size) * 0.1
        self.b2 = np.zeros((1, output_size))
        
        self.lr = lr

    # sigmoid activation
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    
    # derivative of sigmoid
    def sigmoid_derivative(self, a):
        return a * (1 - a)

    # forward pass
    def forward(self, X):
        self.z1 = np.dot(X, self.W1) + self.b1
        self.a1 = self.sigmoid(self.z1)
        
        self.z2 = np.dot(self.a1, self.W2) + self.b2
        self.a2 = self.sigmoid(self.z2)
        
        return self.a2

    # binary cross-entropy loss
    def loss(self, y, y_pred):
        epsilon = 1e-8
        return -np.mean(y * np.log(y_pred + epsilon) + (1 - y) * np.log(1 - y_pred + epsilon))

    # backward pass
    def backward(self, X, y, y_pred):
        m = X.shape[0]

        # output layer error
        dz2 = y_pred - y
        dW2 = np.dot(self.a1.T, dz2) / m
        db2 = np.sum(dz2, axis=0, keepdims=True) / m

        # hidden layer error
        dz1 = np.dot(dz2, self.W2.T) * self.sigmoid_derivative(self.a1)
        dW1 = np.dot(X.T, dz1) / m
        db1 = np.sum(dz1, axis=0, keepdims=True) / m

        # update parameters
        self.W2 -= self.lr * dW2
        self.b2 -= self.lr * db2
        self.W1 -= self.lr * dW1
        self.b1 -= self.lr * db1

    # training
    def fit(self, X, y, epochs=10000):
        for epoch in range(epochs):
            y_pred = self.forward(X)
            loss = self.loss(y, y_pred)
            self.backward(X, y, y_pred)

            if epoch % 1000 == 0:
                print(f"Epoch {epoch}, Loss: {loss:.4f}")

    # prediction
    def predict(self, X):
        y_pred = self.forward(X)
        return (y_pred > 0.5).astype(int)
```

# Example Dataset: XOR
XOR is a classic example because a single linear model cannot solve it well, but a neural network can.

```python
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([
    [0],
    [0],
    [0],
    [1]
])
```

# Train the model

```python
nn = NeuralNetwork(input_size=2, hidden_size=4, output_size=1, lr=0.1)
nn.fit(X, y, epochs=10000)
```

# Test Prediction

```python
predictions = nn.predict(X)
print("Predictions:")
print(predictions)

print("Actual:")
print(y)
```

# What Each part means
## Forward pass
The network computes:
z1 = XW1 + b1
a1 = sigmoid(z1)

z2 = a1W2 + b2
a2 = sigmoid(z2)

- z1 = hidden layer linear combination

- a1 = hidden layer activation

- z2 = output layer linear combination

- a2 = final prediction

## Loss
L = -mean(y*log(y_pred) + (1-y)*log(1-y_pred))
This measures how wrong the predictions are.

## Backpropagation
we calculate gradients:
- how much each weight caused the error
- then subtract a small amount using gradient descent

This is how the network learns.

## Why this works
A neural network is basically:
- weighted sums
- activation functions
- repeated layer by layer
- gradients to update weights

so it is just extension of logistic regression.

Logistic regression:
- one layer only

Neural network:
- hidden layers added


## Very small intuition
Think of it like this:
- weights = what the model believes is important
- forward pass= make prediction
- loss= measure mistake
- backward pass= find who caused the mistake
- update= correct the weights

This is the full learning cycle.
