```python
import numpy as np
import matplotlib.pyplot as plt
```

# Creating some random sample

```python
np.random.seed(20)
x=3*np.random.rand(100,1)
y=4+3*x+np.random.rand(100,1)
```

```python
x.shape
```

```python
y.shape
```

```python
x[2]
```

```python
y[2]
```

# Initializing parameters

```python
w=0
b=0
learning_rate=0.1
epochs=100
n=len(x)
```

# Training loop

```python
for epoch in range(epochs):
    y_pred=w*x+b

    # Computing gradient
    dL_dw=-2/n*np.sum(x*(y-w*x-b))
    dL_db=-2/n*np.sum(y-w*x-b)

    # update the parameter
    w=w-learning_rate*dL_dw
    b=b-learning_rate*dL_db

    if epoch%10==0:
        loss=np.mean((y-y_pred)**2)
        print(f"Epoch {epoch},Loss:{loss}")
print("Final w:",w)
print("Final b:",b)



    
    
```

# Visualizing the result

```python
plt.scatter(x,y)
plt.plot(x,w*x+b,color='red')
plt.show()
```
